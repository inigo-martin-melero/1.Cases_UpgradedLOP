This file contains the cases used in the simulations with 30 variables for 7, 11, 14, 18 and 21 criteria. Each exercise counts with the variables *ranking_table*, *available_budget*, *cost_matrix_client* with the following format.

- *ranking_table*: Python dictionary containing the different clients as keys and an array of positions of each client in each criteria as values.  

- *available_budget*: Python Int representing the maximum budget available to be spent for improvements.   

- *cost_matrix_client*: Python dictionary containing the targeted client as key and a second dictionary as value. This second dictionary is composed of tuples representing (criteria, position t - 1) as keys and the improvement costs as values. The "forbidden movements", like worsening one position, have been disabled by configuring a budget greater than the available one e.g. *available_budget + 1*.
